## 1.1. Handle Values

```python
df.sort_values(by=['ID']).head(10)
```

- Through initial observation, we noticed that there are many duplicates in the 'ID' column. For the same observation ID, we found many additional values

```python
# Function to combine rows with the same ID
def combine_rows(df):
    # Group by 'ID'
    grouped = df.groupby('ID')
    
    # Define a function to handle the combination logic for each group
    def combine_group(group):
        combined_row = {}
        for column in group.columns:
            if column == 'ID':
                combined_row[column] = group[column].iloc[0]
            else:
                unique_values = group[column].dropna().unique()
                if len(unique_values) == 1:
                    combined_row[column] = unique_values[0]
                elif len(unique_values) > 1:
                    # Choose the maximum value in case of having 2 different values of column in same ID
                    combined_row[column] = group[column].max()
                else:
                    combined_row[column] = None
        return pd.Series(combined_row)

    # Apply the combination logic to each group
    combined_df = grouped.apply(combine_group).reset_index(drop=True)
    
    return combined_df

# Combine the rows
combined_df = combine_rows(df)

combined_df.sort_values(by=['ID']).head(10)
```

## 2.1. Encoding

- Transform categorical features to binary features

```python
# Encoding: Transform categorical feautures to binary features
list_encoding = ['Academic_Level','Gender', 'Marital_Status', 'Num_Child']
df_model_encoding = pd.get_dummies(combined_df, columns = list_encoding, dtype=int)
```


## 2.2. Normalization

- Transform all features to the same range
- We have noticed that the dataset contains many outliers. Therefore, we have decided to use MinMaxScaler instead of StandardScaler to mitigate the influence of these outliers.

```python
from sklearn.preprocessing import MinMaxScaler
non_numeric_cols = df_model_encoding.select_dtypes(include=['object', 'datetime']).columns
df_numeric = df_model_encoding.drop(columns=non_numeric_cols)

# Apply MinMaxScaler to the numeric columns
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df_numeric)

# Create a DataFrame with the scaled data
df_scaled = pd.DataFrame(scaled_data, columns=df_numeric.columns)
df_non_numeric = df_model_encoding[non_numeric_cols].reset_index(drop=True)
df_final = pd.concat([df_scaled, df_non_numeric], axis=1)
```

# 3. Dimension Reduction: Principle Components Analysis (PCA)

- In reality, the use of PCA to reduce the number of features for large datasets, is not necessary for this particular case. However, we consider this a case study to have an opportunity to apply this method to larger datasets, so we will still proceed with this approach.

```python
pca = PCA(n_components=3)
pca.fit(df_final)
PCA_ds = pd.DataFrame(pca.transform(df_final), columns=(["PCA1","PCA2", "PCA3"]))
PCA_ds
```

```python
# Proportion of variance explained by each principal component
pca.explained_variance_ratio_
```

# 4. Model Selection

- Choosing K-Means over Hierarchical Clustering is advantageous for high-dimensional data, control over the number of clusters, and ease of interpretation.

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
km = KMeans(init="k-means++", random_state=15, n_init="auto")
visualizer = KElbowVisualizer(km, k=(2,10))

visualizer.fit(df_final)
visualizer.show()
```

```python
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=15)
predicted_labels = kmeans.fit_predict(PCA_ds)

PCA_ds['Clusters']=predicted_labels
combined_df['Clusters']=predicted_labels
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.colors as mcolors

# Define the colors in hexadecimal format
colors_hex = ['#0B090A', '#161A1D', '#660708', '#A4161A', '#BA181B', '#E5383B', '#B1A7A6', '#D3D3D3', '#F5F3F4', '#FFFFFF']

# Convert hexadecimal colors to RGB
colors_rgb = [mcolors.hex2color(color) for color in colors_hex]

# Create colormap
cmap_name = 'custom_palette'
custom_cmap = mcolors.LinearSegmentedColormap.from_list(cmap_name, colors_rgb)

# Plotting
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

sc = ax.scatter(PCA_ds['PCA1'], PCA_ds['PCA2'], PCA_ds['PCA3'], s=40, c=PCA_ds["Clusters"], marker='o', cmap=custom_cmap)

ax.set_title("The Plot Of The Clusters")
ax.set_xlabel('PCA1')
ax.set_ylabel('PCA2')
ax.set_zlabel('PCA3')
ax.set_facecolor('#FFFFE0')

# Adding color bar
cbar = fig.colorbar(sc)
cbar.set_label('Cluster')

plt.show()
```
